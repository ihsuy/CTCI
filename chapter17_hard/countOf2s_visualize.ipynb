```python
def count2s(n):
    count = 0
    while n != 0:
        if n%10 == 2:
            count+=1
        n //= 10
    return count
```

```python
result = []
for i in range(0, 500):
    result.append(count2s(n = i))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(result)
```
